```python
def somar(a, b):
  return a+b
def calcular_media(n1, n2):
  soma = somar(n1, n2)
  media = soma / 2
  return media
nota1 = float(input("digite a primeira nota: "))
nota2 = float(input("digite a segunda nota: "))

resultado = calcular_media(nota1, nota2)
print(f"a média das notas é: {resultado}")
```
