2️. Conversor de Temperatura

```python
def celsius_para_fahrenheit(c):
    f = (c * 9/5) + 32
    print(f"{c}°C equivalem a {f}°F")
```

```python
celsius_para_fahrenheit(10)
```

4️. Quadrado e Cubo

```python
def calcular_potencias(n):
    quadrado = n ** 2
    cubo = n ** 3
    print(f"O quadrado de {n} é {quadrado} e o cubo é {cubo}")
```

```python
calcular_potencias(3)
```

5️. Contagem Regressiva

```python
def contagem_regressiva(inicio):
    print(f"Iniciando contagem a partir do {inicio}:")
    while inicio >= 1:
        print(inicio)
        inicio -= 1
    print("Acabou")
```

```python
contagem_regressiva(2)
```

6️. Cálculo de Desconto

```python
def calcular_desconto(valor, desconto):
    valor_final = valor * (1 - desconto / 100)
    print(f"Valor inicial: R${valor}")
    print(f"Desconto: {desconto}%")
    print(f"Valor final: R${valor_final}")
```

```python
calcular_desconto(100 , 10)
```

7️. Maior e Menor

```python
def extremos(lista):
    maior = lista[0]
    menor = lista[0]
    for num in lista:
        if num > maior:
            maior = num
        elif num < menor:
            menor = num
    print(f"Lista: {lista}")
    print(f"Maior número: {maior}")
    print(f"Menor número: {menor}")
```

```python
extremos([1, 2, 3, 4, 5])
```

8️. Calculadora Simples

```python
def calculadora(a, b, operacao):
    print(f"Operação escolhida: {operacao}")
    if operacao == "soma":
        print(f"{a} + {b} = {a + b}")
    elif operacao == "subtracao":
        print(f"{a} - {b} = {a - b}")
    elif operacao == "multiplicacao":
        print(f"{a} x {b} = {a * b}")
    elif operacao == "divisao":
        if b != 0:
            print(f"{a} ÷ {b} = {a / b}")
        else:
            print("Erro: divisão por zero!")
    else:
        print("Operação inválida!")
```

```python
calculadora(10, 2, "soma")
```

10. Sistema de Notas

```python
def aluno(nota):
    if nota >= 7:
        situacao = "Aprovado"
    elif nota >= 5:
        situacao = "Recuperação"
    else:
        situacao = "Reprovado"
    print(f"Nota: {nota} → Situação: {situacao}")
```

```python
aluno(6.7)
```
